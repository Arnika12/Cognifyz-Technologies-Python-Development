## Task: Create a Data Visualization Tool
Build a tool that takes a dataset andgenerates interactive visualizations usinglibraries such as Matplotlib, Seaborn, orPlotly. This task will enhance theirunderstanding of data visualization principlesand plotting techniques.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Function to load the dataset
def load_data(filepath):
    if filepath.endswith('.csv'):
        return pd.read_csv(filepath)
    elif filepath.endswith('.xlsx'):
        return pd.read_excel(filepath)
    else:
        raise ValueError("Unsupported file format")

# Function for Matplotlib visualizations
def matplotlib_visualization(data):
    plt.figure(figsize=(10,6))
    data.plot(kind='line')  # Example of a line plot
    plt.title('Matplotlib Line Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Function for Seaborn visualizations
def seaborn_visualization(data):
    sns.set(style="darkgrid")
    plt.figure(figsize=(10,6))
    sns.lineplot(data=data)
    plt.title('Seaborn Line Plot')
    plt.show()

# Function for Plotly visualizations
def plotly_visualization(data):
    fig = px.scatter(data_frame=data, x=data.columns[0], y=data.columns[1], title='Plotly Scatter Plot')
    fig.show()

# Main function to handle the process
def generate_visualization(filepath):
    data = load_data(filepath)
    
    print(f"Data Preview:\n{data.head()}\n")

    # Loop until the user gives a correct input
    while True:
        library = input("Select library to generate visualization ('matplotlib', 'seaborn', 'plotly'): ").lower()
        if library == 'matplotlib':
            matplotlib_visualization(data)
            break
        elif library == 'seaborn':
            seaborn_visualization(data)
            break
        elif library == 'plotly':
            plotly_visualization(data)
            break
        else:
            print("Invalid choice. Please choose between 'matplotlib', 'seaborn', or 'plotly'.")

# Example usage
if __name__ == "__main__":
    filepath = 'iris.csv'  # Specify your dataset path
    generate_visualization(filepath)
```
